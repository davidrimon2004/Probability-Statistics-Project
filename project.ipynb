```python
import statistics as stat
import pandas as pd
import matplotlib.pyplot as visual
import seaborn as sb
```

```python
data ="Diabetes Dataset.csv"
read=pd.read_csv(data)
preg=read[read.columns[0]]
gloc=read[read.columns[1]]
pressure=read[read.columns[2]]
thick=read[read.columns[3]]
insulin=read[read.columns[4]]
bmi=read[read.columns[5]]
diabetes=read[read.columns[6]]
age=read[read.columns[7]]
outcome=read[read.columns[8]]
```

```python
def Scatterplot(x_axis,y_axis,IndX,IndY):
    sb.scatterplot(x=x_axis,y=y_axis)
    visual.xlabel(read.columns[IndX])
    visual.ylabel(read.columns[IndY])
    visual.title('Pressure & BMI')
    visual.show()
Scatterplot(pressure,bmi,2,5)
```

```python
plot=read.corr()
sb.heatmap(plot, annot=True, cmap='plasma', fmt=".2f", cbar=True)
visual.show()
```

```python
def pie(x):
  count=read[read.columns[x]].value_counts()
  visual.pie(count,labels=['0','1'])
  visual.show()
```

```python
pie(8)
```

```python
def boxplot (indX,title):
    read.boxplot(column=read.columns[indX])
    visual.title(title)
    visual.show()
```

```python
def dispersion(x):
  standard=stat.stdev(x)
  print('standard deviation :', standard)
  variance=stat.variance(x)
  print('variance :', variance)
```

```python
for i in range(8):
  dispersion(read[read.columns[i]])
```

```python
def me_me_mo(x):
  print(stat.mean(read[read.columns[x]]))
  print(stat.median(read[read.columns[x]]))
  print(stat.mode(read[read.columns[x]]))
```

```python
for x in range(0,9):
  me_me_mo(x)
  print("======================")
```

```python
def min_max(x):
  print( min(read[read.columns[x]]))
  print( max(read[read.columns[x]]))
```

```python
for x in range(0,8):
  min_max(x)
  print("======================")
```
